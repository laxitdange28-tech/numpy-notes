# Vectorization Kya Hai?

Vectorization ka matlab hai arrays par operations ko without loops perform karna.

NumPy internally C language me optimized hota hai, isliye vectorized operations Python loops se bahut fast hote hain.

# Without Vectorization (Using Loop)

```python
arr = [1,2,3,4,5]
result = []

for i in arr:
  result.append(i*2)

print(result)
```

# With Vectorization (Using NumPy)

```python
import numpy as np

arr = np.array([1,2,3,4,5])
result = arr*2

print(result) # NumPy ne har element par automatically multiplication kar diya:
```

# Vectorization ke Benefits

| Benefit          | Description               |
| ---------------- | ------------------------- |
| Faster           | Loops se bahut fast       |
| Less Code        | Kam code likhna padta hai |
| Easy Readability | Code clean dikhta hai     |
| Optimized        | C backend use karta hai   |


# Broadcasting Kya Hai?

Broadcasting ek technique hai jisme NumPy different shapes ke arrays par arithmetic operations perform kar sakta hai.

Agar dimensions compatible ho to NumPy automatically smaller array ko expand kar deta hai.

# Scalar Broadcasting

```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr+10)
```

# Row Bordcasting

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print(arr  +  np.array([10,20,30]))
```

# Vectorization VS Broadcating

| Feature | Vectorization           | Broadcasting                  |
| ------- | ----------------------- | ----------------------------- |
| Purpose | Loops remove karna      | Different shapes handle karna |
| Speed   | Fast                    | Fast                          |
| Example | arr * 2                 | arr + scalar                  |
| Use     | Element-wise operations | Shape expansion               |
