##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np
```

```python
arr = np.array([10,20,30,40,50])
# pick elements at position 0,2,4

print(arr[[0,2,3]])
```

You can do this fancy indexing in 2D Matrix aswell

```python
arr = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(arr[[0,1,2],[0,1,2]])
```

```python
matarix = np.arange(1,13).reshape(3,4)
print(matarix)
print(matarix[[0,1],[1,2]])
```

```python
print(matarix[[1,2],[0,2]])
```

##Conditional Selection
Find values based on conditions or replace them.

```python
arr = np.array([5,7,11,18,30])

#we will get the indices only
# which are greter than 10
print(np.where(arr>10))

# array me jaha bhi 10 se choti values
#aa rhi hai waha 0 se replace kar do
print(np.where(arr>10,arr,0))
```

```python
arr2 = np.array([23,35,30,50,100,75,80])
print(np.where(arr2>55,arr2,0))
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
#Aam Zindagi
arr = np.array([2,10,56,1,54])
print(np.sort(arr))
```

```python
# Matrix sort
arr = np.array([[2,10,56],[1,54,23]])
# row wise sort (axis = 1)
print(np.sort(arr,axis=1))
# column wise sort (axis = 0)
print(np.sort(arr,axis=0))
#sort complete matrix
np.sort(arr)
```

```python
index = np.array([2,10,56,1,54])
print(np.argsort(index))
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.concatenate([a,b]))
```

Vertical & Horizontal Stacking

```python
arr = np.array([[1,2,3],[4,5,6]])
# vertical staking
arr2 = np.array([[7,8,9],[10,11,12]])
print(np.vstack([arr,arr2]))
```

```python
arr = np.array([[1,2,3],[4,5,6]])
# horizontal staking
arr2 = np.array([[7,8,9],[10,11,12]])
print(np.hstack([arr,arr2]))
```

```python
arr = np.array([[1,2,3],[4,5,6]])

for i in arr:
  for j in i:
    print(j)
```
