# Q1. Fancy Indexing (1D Array)

## Problem Statement

```python
[10, 20, 30, 40, 50, 60]
```

- 10
- 30
- 60

## Sample Output
```python

```


# Q5. Conditional Selection

## Problem Statement
Create the array:

```python
[2, 4, 7, 10, 13, 16]
```

Print only even numbers.

## Sample Output
```python
[ 2  4 10 16]
```


````python
# Write your code here
arr = np.array([2,4,7,10,13,16])
print(arr[arr % 2==0])
````

```python
arr1 = [1,2,3]

```

## Sample Output


# Q9. Vertical Stacking

## Problem Statement
Create two arrays:

```python
[1,2,3]
[4,5,6]
```

Combine them using vertical stacking.

## Sample Output
```python
[[1 2 3]
 [4 5 6]]
```


````python
# Write your code here
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(np.vstack((arr1,arr2)))
````


# Q10. Horizontal Stacking

## Problem Statement
Create two arrays:

```python
[10,20,30]
[40,50,60]
```

Combine them using horizontal stacking.

## Sample Output
```python
[10 20 30 40 50 60]
```


````python
# Write your code here
arr1 = np.array([10,20,30])
arr2 = np.array([40,50,60])

print(np.hstack((arr1,arr2)))
````
