
# Q10. Concatenation and Stacking

## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
# Write your code here

import numpy as np

# Create arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Concatenate arrays
concat_arr = np.concatenate((arr1, arr2))

# Vertical Stack
vstack_arr = np.vstack((arr1, arr2))

print("Concatenated Array:")
print(concat_arr)

print("\nVertically Stacked Array:")
print(vstack_arr)
````
